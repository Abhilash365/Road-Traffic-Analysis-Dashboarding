```python
import pandas as pd
import requests



def get_weather(lat, lon, crash_date):
    try:
        # Handle missing crash_date safely
        if pd.isna(crash_date):
            return None
        
        # Format date as yyyy-mm-dd
        date_str = pd.to_datetime(crash_date, errors="coerce").strftime("%Y-%m-%d")
        if date_str is None:
            return None
        
        # Build API URL
        url = (
            f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/"
            f"{lat},{lon}/{date_str}?unitGroup=metric&include=days&key={API_KEY}"
        )
        
        # Call API
        response = requests.get(url)
        response.raise_for_status()
        
        data = response.json()
        
        # Extract first day's weather if available
        if "days" in data and len(data["days"]) > 0:
            return data["days"][0]
        else:
            return None

    except Exception as e:
        print(f"âŒ Error for lat={lat}, lon={lon}, date={crash_date}: {e}")
        return None
```

```python
from geopy.geocoders import Nominatim
from geopy.distance import geodesic

def nearest_hospital_distance(lat, lon, limit=5):
    """
    Find the nearest hospital distance (in km) from given latitude and longitude.
    Uses OpenStreetMap Nominatim API.
    """
    my_location = (lat, lon)
    geolocator = Nominatim(user_agent="hospital_locator", timeout=10)

    # search hospitals near the given coordinates
    hospitals = geolocator.geocode(
        f"hospital near {lat}, {lon}",
        exactly_one=False,
        limit=limit
    )

    if not hospitals:
        return None  # no hospitals found

    # find the nearest hospital
    nearest = None
    min_distance = float("inf")

    for h in hospitals:
        hosp_location = (h.latitude, h.longitude)
        dist = geodesic(my_location, hosp_location).km
        if dist < min_distance:
            min_distance = dist
            nearest = h

    return min_distance


```

```python
import pandas as pd

# --- Step 1: Define national holidays for each year ---
national_holidays = {
    2021: [
        "2021-01-26", "2021-03-29", "2021-04-02", "2021-04-21", "2021-04-25",
        "2021-05-14", "2021-05-26", "2021-07-21", "2021-08-15", "2021-08-19",
        "2021-08-30", "2021-10-02", "2021-10-15", "2021-10-19", "2021-11-04", "2021-11-19"
    ],
    2022: [
        "2022-01-26", "2022-03-01", "2022-03-18", "2022-04-14", "2022-04-15",
        "2022-05-03", "2022-05-16", "2022-07-10", "2022-08-15", "2022-08-19",
        "2022-10-02", "2022-10-05", "2022-10-09", "2022-10-24", "2022-11-08", "2022-12-25"
    ],
    2023: [
        "2023-01-26", "2023-03-08", "2023-03-30", "2023-04-04", "2023-04-07",
        "2023-04-22", "2023-05-05", "2023-06-29", "2023-07-29", "2023-08-15",
        "2023-09-07", "2023-09-28", "2023-10-02", "2023-10-24", "2023-11-12", "2023-11-27", "2023-12-25"
    ]
}

# --- Step 2: Generate weekends and combine with national holidays ---
all_dates = []

for year in [2021, 2022, 2023]:
    # Generate weekends (Saturday=5, Sunday=6)
    weekends = pd.date_range(f"{year}-01-01", f"{year}-12-31", freq="D")
    weekends = weekends[weekends.weekday >= 5]
    
    # Convert national holidays to datetime
    nh = pd.to_datetime(national_holidays[year])
    
    # Combine weekends + national holidays
    combined = pd.Series(pd.concat([weekends.to_series(), nh.to_series()])).drop_duplicates()
    
    # Add to all years
    all_dates.extend(combined)

# --- Step 3: Create DataFrame ---
df = pd.DataFrame({"Date": pd.to_datetime(all_dates).sort_values().unique()})
df["Type"] = df["Date"].apply(
    lambda d: "Weekend" if d.weekday() >= 5 else "National Holiday"
)

# --- Step 4: Save to CSV ---
df.to_excel("india_holidays_2021_2023.xlsx", index=False)

print("âœ… CSV generated: india_holidays_2021_2023.csv")
print(df.head(10))
```
